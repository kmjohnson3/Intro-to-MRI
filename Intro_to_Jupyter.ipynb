```
# In Python you import libraries to use

# Numpy is an array library 
import numpy as np

# Matplotlib is a plotting library
import matplotlib.pyplot as plt

# These are core pyhton libaries
import math
```

```
# Make a simple array 
t = np.linspace(start=0, stop=2*math.pi, num=100, dtype=np.float32)
y = np.sin(t)
```

```
# Simple plot
plt.figure()
plt.plot(t,y)
plt.xlabel('Time')
plt.ylabel('Signal')
plt.show()
```
